```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Define the file path
file_path = '/content/drive/Shareddrives/DATA 270 GWAR/GWAR_dataset.csv'  # Replace with the actual file path

# Load the dataset
data = pd.read_csv(file_path)

# Reduce the size of the dataset
fraction = 0.5  # Adjust this value to change the fraction of data you want to use
data = data.sample(frac=fraction, random_state=42)

# Define feature and target columns
feature_cols = ['Caffeine', 'Carbohydrate', 'FAT', 'Fat, saturated', 'Fiber', 'Iron', 'Lactose', 'Protein', 'Starch',
          'TOTAL TRANS FATTY ACID', 'Vitamin A', 'Vitamin B12', 'Vitamin C', 'Vitamin D', 'Vitamin E']
target_col = ['CALORIES']

# Split the data into features and target
X = data[feature_cols]
y = data[target_col]

# Preprocessing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y.values.reshape(-1, 1))

# Split the data into train, validation, and test sets
X_train_val, X_test, y_train_val, y_test = train_test_split(X_scaled, y_scaled, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.17647, random_state=42)

# Train the SVR model
model = SVR(kernel='linear')
model.fit(X_train, y_train.ravel())

# Evaluate on the training set
y_pred_train = model.predict(X_train)
train_mse = mean_squared_error(y_train, y_pred_train)
train_rmse = np.sqrt(train_mse)
train_mae = mean_absolute_error(y_train, y_pred_train)
train_r2 = r2_score(y_train, y_pred_train)
print('Train MSE:', train_mse)
print('Train RMSE:', train_rmse)
print('Train MAE:', train_mae)
print('Train R2:', train_r2)

# Validate the model
y_pred_val = model.predict(X_val)
val_mse = mean_squared_error(y_val, y_pred_val)
val_rmse = np.sqrt(val_mse)
val_mae = mean_absolute_error(y_val, y_pred_val)
val_r2 = r2_score(y_val, y_pred_val)
print('Validation MSE:', val_mse)
print('Validation RMSE:', val_rmse)
print('Validation MAE:', val_mae)
print('Validation R2:', val_r2)

# Test the model
y_pred_test = model.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred_test)
test_rmse = np.sqrt(test_mse)
test_mae = mean_absolute_error(y_test, y_pred_test)
test_r2 = r2_score(y_test, y_pred_test)
print('Test MSE:', test_mse)
print('Test RMSE:', test_rmse)
print('Test MAE:', test_mae)
print('Test R2:', test_r2)
```
